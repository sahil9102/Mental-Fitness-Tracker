```python
import pandas as pd
import pickle
```

```python
df1 = pd.read_csv("mental-and-substance-use-as-share-of-disease.csv")
df2 = pd.read_csv("prevalence-by-mental-and-substance-use-disorder.csv")
data = pd.merge(df2,df1)
```

```python
data.isnull().sum()
```

```python
data.drop("Code",axis=1,inplace=True)
data.isnull().sum()
```

```python
data = data.set_axis(['Country','Year','Schizophrenia', 'Bipolar_disorder', 'Eating_disorder','Anxiety','Drug_usage','Depression','Alcohol','Mental_fitness'], axis='columns', copy=False)
data
```

```python
data = pd.DataFrame(data)
```

```python
pickle.dump(data,open("merged_data.pkl","wb"))
```
